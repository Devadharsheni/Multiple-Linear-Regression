```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="/content/Student_Performance.csv"
df=pd.read_csv(path)
df.head(5)
```

```python

sns.jointplot(x='Hours Studied',y='Performance Index',data=df,alpha=0.5)
```

```python
sns.pairplot(df,
             kind='scatter',
             plot_kws={'alpha':0.4},
             diag_kws={'alpha':0.55,'bins':40})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']]
y=df['Performance Index']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print(x_train.shape)
```

```python
from sklearn.linear_model import  LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
y_pred=lr.predict(x_test)
```

```python
print(y_pred)
```

```python
sns.scatterplot(x=y_test,y=y_pred)
plt.ylabel('Predicted Performance Index')
plt.xlabel('Actual value')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
import math
```

```python
residual=y_test-y_pred
print(residual)
```

```python
sns.displot(residual,bins=40)
```
